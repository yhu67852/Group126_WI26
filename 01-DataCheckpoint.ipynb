## Research Question

Is there a statistically significant relationship between recent win/loss outcomes and continuous play session length in competitive Marvel Rivals players? Additionally, can match outcome patterns be used to predict when a player is likely to stop playing?

## Background and Prior Work

Competitive online video games such as Marvel Rivals, a team-based first-person-shooter, provide a natural setting to study how reward and failure influence persistence. Wins and losses are not merely outcomes of play; they engage reinforcement learning mechanisms supported by dopaminergic reward signaling. A common misconception of dopamine is that dopamine neurons always activate upon receiving positive rewards. However, research in Reward Prediction Error (RPE) shows that dopamine neurons don't just respond to any reward: they respond only to rewards that differ from their prediction.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Take food, for example: you realize the dining hall just added your favourite food to the menu, so you ecstatically purchase it. Perhaps the next day, you'll get it again. But if you continue to buy it every single day, for 3 meals a day, for an entire month—you'll probably be sick of it, despite it having been your favourite food. The same applies in games—if you only ever win, at some point you'll get sick of it and quit. 

RPE suggests that unexpected outcomes can shape motivation and prolong engagement, but that still requires the right amount of variability in reinforcement schedules. Just as winning too much can dull your gameplay experience, losing too much could deter you from wanting to experience more. Most players believe that competitive MMR (Matchmaking Rating) systems aim for players to have roughly a 50% win/loss ratio, as "it means it's where you should be," (in terms of rank and skill) but some have also noted that it also creates negative feelings caused by being "hardstuck," <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). a phenomenon where players reach a bottleneck in terms of win/loss ratio, resulting in stagnant rankings that can contribute to feelings of demotivation. Hence, there is reason to suspect that MMR systems are not so transparently "50/50."

Prior works show that there is a temporal attractiveness in uncertainty and resolution—uncertainty is worth it because resolution awaits<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). Our desire to move from a state of "unknown" to "knowing" is in itself valuable, regardless of whether the outcome is positive or negative. This added value of uncertainty is coined "resolution utility," (Ruan et al., 2018, Shen et al., 2019) <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3), and is crucial to why player's don't "quit halfway." The result of this research not only provides reasons for why continuous gameplay occurs, it also helps validate that a player's decision to stop playing is determined by what happens after resolution, rather than issues arising during gameplay.

We specifically chose competitive Marvel Rivals because the added layer of social rewards in tandem with wins/losses will provide a stronger reinforcement effect. Competitive Marvel Rivals segregates players by skill, resulting in increasing game difficulty as the player improves, thereby ensuring that reward outcomes continue to be variable in nature<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). The rank labels displayed on player accounts introduce social status and evaluation, which can increase motivation to continue playing and make the decision to stop a more consequential choice. 

We recognize that ending a gaming session may come from a variety of reasons: fatigue, obligations (e.g., work or school), social factors, technical interruptions, and more. However, this project focuses on whether match outcome patterns have a statistically meaningful relationship with variation in play duration, which, if found, would indicate cause beyond these background factors. 

In summary, this project asks whether win–loss patterns correlate with continuous play duration in competitive Marvel Rivals. If such relationship is found, we then ask whether we are able to predict a stopping point for player gaming sessions.

## Hypothesis


We believe that there is a strong relationship between win/loss outcomes and continuous play duration in competitive Marvel Rivals. We predict that players will tend to play longer after losses or alternating win–loss streaks compared to consistent winning outcomes, reflecting loss aversion and variable reinforcement effects. Because ranked games involve both dopaminergic reward feedback and social motivation through visible rank, we expect these outcome patterns to increase player persistence within a session. 

However, we also predict that extended loss streaks (e.g., beyond ~5 consecutive losses) will be associated with session termination, as sustained negative outcomes reduce rewarding feedback and may increase frustration, fatigue, and disengagement over time.

### Dataset #1: Match History of Top 984 Ranked Marvel Rivals Players
#### Overview
This dataset contains detailed match history data for the current top 984 ranked Marvel Rivals players. It includes 71,541 total matches and 14 variables describing match-level and player-level information. Each row represents a single match from the perspective of a specific player. The dataset was constructed by aggregating match history records for publicly visible top-ranked players. Player identifiers are anonymized before analysis.

##### Important Metrics
1. **match_play_duration (seconds, float):**
This variable represents the total duration of a match in seconds. Shorter matches may reflect early forfeits or dominant victories, while longer matches may indicate closely contested games or specific game modes. Also important for identifying continuous gameplay and potential fatigue.

2. **match_time_stamp (datetime):**
Originally recorded as Unix time, converted to datetime format. This indicates when the match was played and allows for identification of continuous gameplay.

3. **is_win (boolean):** Indicates whether the player won the match. This variable allows us to evaluate win/loss streak effects.

4. **add_score (float):** Float value indicating rank rating (RR) gains/loss. RR gain and loss is directly correlated with match outcome, however differences in magnitude may have effects on the impact of the win/loss.

##### Data Cleaning

The dataset contains no major missing values in our critical fields. Numeric columns were examined for abnormal values. Some suspicious values were detected (e.g., match_winner_side had values up to 12 before cleaning), and these were filtered to retain only valid binary outcomes (0 or 1). Columns with excessive missingness (>50%) were removed.

##### Concerns and Limitations
1. Elite Players (Sampling Bias): This dataset includes only the top ~900 ranked players. As a result, lower-ranked and casual players are not represented, meaning behavioral patterns may differ significantly from the broader population (as most top players tend to be professional and could have other reasons for gameplay). Top players also play more frequently, have longer streaks, and be more strategically motivated. This introduces selection bias and limits generalizability.
2. Public Profiles: Data was collected from publicly accessible player profiles. Players who restrict visibility or are not on the leaderboard are excluded. This further biases the dataset toward: Highly active, competitive, and potentially well-known players.
3. API Data Constraints: The data was collected via API access that requires player UIDs. This limits sampling randomness and may introduce non-random selection effects, incomplete match histories, and possibly unknown filtering criteria imposed by the API.
4. Temporal Concentration: Most matches occurred within a narrow timeframe. If the data spans only part of a season, conclusions may reflect short-term game/meta trends rather than long-term patterns.

```python
import json
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

raw_path = './data/00-raw/marvel_rivals_raw.json'
with open(raw_path, 'r') as f:
    raw_data = json.load(f)

all_matches = []

for player, player_data in raw_data.items():
    matches = player_data.get('match_history', [])
    for match in matches:
        match['player'] = player
        all_matches.append(match)

df = pd.DataFrame(all_matches)

if 'match_time_stamp' in df.columns:
    df['match_time_stamp'] = pd.to_datetime(df['match_time_stamp'], unit='s')

print(f"Number of matches: {len(df)}")
print(f"Number of columns: {df.shape[1]}")
print("\n\nInitial Dataframe")
print(df.head())

print("\nDataFrame info:\n")
print(df.info())

missing_counts = df.isna().sum()
missing_percent = df.isna().mean() * 100
missing_summary = pd.DataFrame({'missing_count': missing_counts, 'missing_percent': missing_percent})
print("\nMissing data summary:\n", missing_summary)

#missingness bar chart
plt.figure(figsize=(10,4))
missing_percent.sort_values(ascending=False).plot(kind='bar')
plt.ylabel("Percent Missing")
plt.title("Missingness by Column (%)")
plt.tight_layout()
plt.show()

#drop map thumbnails (many missing & irrelevant to our data)
if 'map_thumbnail' in df.columns:
    df = df.drop(columns=['map_thumbnail'])
    print("NOTE: Dropped map_thumbnail (non-analytical display field)")

#missingness heatmap
plt.figure(figsize=(12,6))
sns.heatmap(df.isna(), cbar=False)
plt.title("Missing Data Heatmap")
plt.show()

numeric_cols = df.select_dtypes(include='number').columns
print("\nNumeric column statistics:\n")
print(df[numeric_cols].describe())

for col in numeric_cols:
    if (df[col] < 0).any():
        print(f"Suspicious negative values in column: {col}")

critical_cols = ['player', 'match_time_stamp', 'match_play_duration']
df_clean = df.dropna(subset=critical_cols)

df_clean[numeric_cols] = df_clean[numeric_cols].fillna(0)

threshold = 0.5
df_clean = df_clean.loc[:, df_clean.isna().mean() < threshold]

if 'match_winner_side' in df_clean.columns:
    df_clean = df_clean[df_clean['match_winner_side'].isin([0, 1])]

#tidy up extra info in the match_player dictionary
for col in ['match_player', 'score_info']:
    if col in df_clean.columns:
        expanded = df_clean[col].apply(lambda x: x if isinstance(x, dict) else {}).apply(pd.Series)
        expanded = expanded.add_prefix(f"{col}_")
        df_clean = pd.concat([df_clean.drop(columns=[col]), expanded], axis=1)

numeric_cols_clean = df_clean.select_dtypes(include='number').columns

print(f"\nAfter cleaning/tidying: {len(df_clean)} rows, {df_clean.shape[1]} columns")
print(df_clean.head())

os.makedirs('./data/01-interim', exist_ok=True)
df_clean.to_json('./data/01-interim/marvel_rivals_interim.json', orient='records', indent=4)

os.makedirs('./data/02-processed', exist_ok=True)
df_clean.to_json('./data/02-processed/marvel_rivals_clean.json', orient='records', indent=4)
df_clean.to_csv('./data/02-processed/marvel_rivals_clean.csv', index=False)
```

```python
#plot match duration variability
plt.figure(figsize=(8,5))
sns.histplot(df_reduced["match_play_duration"], bins=40)
plt.xlabel("Match Duration (seconds)")
plt.title("Distribution of Match Duration")
plt.show()

#plot win/loss variability
sns.countplot(x="is_win", data=df_reduced)
plt.title("Win vs Loss Distribution")
plt.show()

#plot matches per player
matches_per_player = df_reduced["anon_player_id"].value_counts()

plt.figure(figsize=(8,5))
sns.histplot(matches_per_player, bins=40)
plt.xlabel("Matches per Player")
plt.title("Distribution of Match Volume per Player")
plt.show()


#plot time between matches but trim large gaps

df_sorted = df_reduced.sort_values(["anon_player_id", "match_time_stamp"]).copy()

df_sorted["match_end_time"] = (
    df_sorted["match_time_stamp"] +
    pd.to_timedelta(df_sorted["match_play_duration"], unit="s")
)

df_sorted["next_match_start"] = (
    df_sorted.groupby("anon_player_id")["match_time_stamp"].shift(-1)
)

df_sorted["idle_gap_sec"] = (
    df_sorted["next_match_start"] - df_sorted["match_end_time"]
).dt.total_seconds()

df_sorted["idle_gap_sec"] = df_sorted["idle_gap_sec"].clip(lower=0)

df_sorted["idle_gap_min"] = df_sorted["idle_gap_sec"] / 60

gap_data = df_sorted["idle_gap_min"].dropna()

print("Summary statistics BEFORE trimming (minutes):")
print(gap_data.describe())

threshold_min = 120  #seconds

extreme_count = (gap_data > threshold_min).sum()
total_count = len(gap_data)

print(f"\nDropping {extreme_count} gaps (> {threshold_min} minutes) for visualization clarity")
print(f"That is {extreme_count / total_count:.2%} of all computed gaps")

gap_trimmed = gap_data[gap_data <= threshold_min]

print("\nSummary statistics AFTER trimming (minutes):")
print(gap_trimmed.describe())

# Plot trimmed distribution
plt.figure(figsize=(8,5))
sns.histplot(gap_trimmed, bins=100)
plt.xlabel("Idle Time Between Matches (minutes)")
plt.title(f"Idle Time Between Matches (≤ {threshold_min} Minutes)")
plt.show()

#adding new categorical metric to predict for data processing
#create a new column in the dataframe called match order, which is the chronological order of the matches by person
df_sorted[ "match_order" ] = (df_sorted.set_index( "match_time_stamp" ).groupby( "anon_player_id" ).cumcount() + 1).values

#creating match before column to create a merging index
df_sorted[ "match_before" ] = df_sorted[ "match_order" ] - 1

#using merge to put the "idle_gap_min" column from the previous match chronologically into the present
#should have minutes from previous match to current and minutes from current to next in the same column now
temp_df = pd.merge( df_sorted, df_sorted[["anon_player_id", "idle_gap_min", "match_order"]], 
                   left_on = ["anon_player_id","match_before"], 
                   right_on = ["anon_player_id", "match_order"], how = "left",
                   suffixes=('_match_current', '_match_before')
                   ).drop("match_order_match_before", axis = 1 ).rename( {"match_order_match_current" : "match_order",
                                                                         "idle_gap_min_match_current" : "idle_gap_min" }, axis = 1 )

#fill empty values in the minute columns for easier processing
temp_df = temp_df.fillna({
    "idle_gap_min_match_before": -1, 
    "idle_gap_min": -1 })

#assignment function for categorical variable to be predicting
# "end series" means match within 45 before but no match within 45 after
# "independent match" means no match within 45 before or after
# "in series" means match within 45 after
def serialization(row):
    before = row["idle_gap_min_match_before"]
    current = row["idle_gap_min"]
    if int(round(current)) == -1 & int(round(before)) < 45:
        return "end series"
    elif int(round(current)) == -1 & int(round(before)) >= 45:
        return "independent match"
    elif int(round(current)) >= 45 & int(round(before)) == -1:
        return "independent match"
    elif int(round(current)) < 45:
        return "in series"
    elif int(round(current)) >= 45 & int(round(before)) < 45:
        return "end series"
    elif int(round(current)) >= 45 & int(round(before)) >= 45:
        return "independent match"
    else:
        return numpy.NAN

#create categorical column using previous function
temp_df["series_categorical"] = temp_df.apply( serialization, axis = 1 )

#make sure no failures to assign
print(temp_df["series_categorical"].isna().sum())

#visualizations
#Is win
#we can eyeball that it looks like proportionally more matches that ended series were losses than those in series
series_win = sns.histplot( temp_df, x = "series_categorical", hue = "is_win" )
plt.xlabel("Count of categories")
plt.ylabel("Series Categories")
plt.show()

#Change in Score
#There isn't an obvious difference in the distribuitons between in and end of series
score_series = sns.histplot( temp_df, x = "add_score", hue = "series_categorical" )
plt.xlabel("Score Values")
plt.show()

#Duration
#similar to the previous, the distributions look fairly equal
duration_series = sns.histplot( temp_df, x = "match_play_duration", hue = "series_categorical" )
plt.xlabel("Play Duration")
plt.show()
```
