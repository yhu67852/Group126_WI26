# Research Question

Among the top 1000 Marvel Rivals players, to what extent can recent negative performance signals—such as loss streaks, rating loss magnitude (SR change), and poor KDA—predict the end of a continuous play session?

## Background and Prior Work

Competitive online video games such as Marvel Rivals, a team-based first-person-shooter, provide a natural setting to study how reward and failure influence persistence. Performance statistics are not merely outcomes of play; they engage reinforcement learning mechanisms supported by dopaminergic reward signaling. A common misconception of dopamine is that dopamine neurons always activate upon receiving positive rewards. However, research in Reward Prediction Error (RPE) shows that dopamine neurons don't just respond to any reward: they respond only to rewards that differ from their prediction.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Unexpected negative outcomes, such as losing when one expected to win, can therefore generate strong motivational signals that influence subsequent behavior.

In competitive games, performance feedback is not limited to binary win–loss outcomes. Players also receive more nuanced signals about their performance, including rating changes (such as Skill Rating or SR gains and losses) and individual performance metrics such as kills, deaths, and assists (KDA). These signals provide continuous feedback about whether a player is performing well relative to their expectations. Negative performance signals—such as loss streaks, large rating losses, or poor KDA—may therefore influence whether players continue playing or end their play session.

We focus specifically on high-ranking Marvel Rivals players on the Top 1000 Leaderboard. Because elite players invest substantial time and effort into achieving and maintaining their rank, they are likely to be highly engaged with competitive play and may participate in longer, more sustained gameplay sessions. As a result, their decision to stop playing may be more strongly tied to performance feedback—such as loss streaks, rating losses, or poor in-game performance—rather than external constraints alone. Although this population represents only a small subset of the overall player base, it provides a useful setting for examining how competitive performance signals influence persistence in gameplay.

Prior work suggests that uncertainty and outcome resolution contribute to continued engagement. The transition from uncertainty to resolution can itself produce psychological value, a concept known as "resolution utility."<a name="cite_ref-2"></a><sup>2</sup> <a name="cite_ref-3"></a><sup>3</sup> In gaming contexts, this mechanism may motivate players to continue playing across multiple matches in order to resolve uncertainty about their performance trajectory. However, once outcomes are resolved—particularly when they are repeatedly negative—players may reassess whether continuing to play is worthwhile.

In summary, this project investigates whether recent negative performance signals—such as loss streaks, rating loss magnitude (SR change), and poor KDA—predict the end of a continuous play session among high-ranking Marvel Rivals players.

# Hypothesis


We hypothesize that negative performance signals will be associated with an increased likelihood of ending a continuous play session among high-ranking Marvel Rivals players. Specifically, we predict that players will be more likely to stop playing following sustained loss streaks, larger decreases in Skill Rating (SR), and matches with poorer individual performance (lower KDA).

However, we expect that isolated negative outcomes will not strongly predict session termination. Because competitive players are highly motivated to recover from short-term setbacks, we anticipate that session endings will be more strongly associated with persistent negative performance trends—such as multiple consecutive losses or consistently poor performance—rather than single unfavorable matches.

#### Section 1 of EDA - Defining Sessions of Continuous Gameplay

To begin our analsysis, we load the fully processed dataset and perform another step of sorting. Because our research question concerns whether negative performance signals predict the end of a continuous play session, we first construct play sessions based on the time gap between matches, then examine distributions of match characteristics.

**Construct Play Sessions**

```python
#plot time between matches but trim large gaps
# sort matches chronologically within each player
df_sorted = df_reduced.sort_values(["anon_player_id", "match_time"]).copy()

# compute end time of each match
df_sorted["match_end_time"] = (
    df_sorted["match_time"] +
    pd.to_timedelta(df_sorted["match_duration"], unit="s")
)

# compute start time of next match for each player
df_sorted["next_match_start"] = (
    df_sorted.groupby("anon_player_id")["match_time"].shift(-1)
)

# idle gap until next match
df_sorted["idle_gap_sec"] = (
    df_sorted["next_match_start"] - df_sorted["match_end_time"]
).dt.total_seconds()

# clip small negative values caused by timing precision
df_sorted["idle_gap_sec"] = df_sorted["idle_gap_sec"].clip(lower=0)

# convert to minutes
df_sorted["idle_gap_min"] = df_sorted["idle_gap_sec"] / 60
```

```python
#Plot idle gap distribution
gap_data = df_sorted["idle_gap_min"].dropna()

print("Summary statistics BEFORE trimming (minutes):")
print(gap_data.describe())

threshold_min = 120  #seconds

extreme_count = (gap_data > threshold_min).sum()
total_count = len(gap_data)

print(f"\nDropping {extreme_count} gaps (> {threshold_min} minutes) for visualization clarity")
print(f"That is {extreme_count / total_count:.2%} of all computed gaps")

gap_trimmed = gap_data[gap_data <= threshold_min]

print("\nSummary statistics AFTER trimming (minutes):")
print(gap_trimmed.describe())
```

```python
# Plot trimmed distribution
plt.figure(figsize=(8,5))
sns.histplot(gap_trimmed, bins=100)
plt.xlabel("Idle Time Between Matches (minutes)")
plt.title(f"Idle Time Between Matches (≤ {threshold_min} Minutes)")
plt.show()
```

This figure shows the distribution of idle time between consecutive matches for each player. While some gaps are extremely large—reflecting players stopping for hours or days—most gaps are relatively short. After trimming gaps larger than 120 minutes for visualization clarity, the median idle time between matches is approximately 10 minutes, with most gaps falling between about 5 and 20 minutes. This pattern suggests that players frequently queue into another match shortly after finishing one, indicating that matches often occur in clusters of continuous play. These observations support the use of a time-based threshold to define play sessions in later analyses.

**Define Sessions**

This section constructs a categorical variable that describes whether each match occurs within a continuous sequence of play or represents the end of a session. First, matches are ordered chronologically for each player to identify their sequence of games. The time gaps between matches are then used to determine whether another match occurred within a specified threshold (45 minutes) before or after the current match. 

Using this information, each match is classified into one of three categories: **“in series”**, meaning it is part of a continuous session of play; **“end series”**, meaning it is the final match before a long break; and **“independent match”**, meaning the match occurs in isolation without nearby matches. This categorization provides a way to approximate player sessions and enables later analysis of whether certain outcomes, such as losses or rating changes, are associated with players stopping their play session.

```python
#create a new column in the dataframe called match order, which is the chronological order of the matches by person
df_sorted[ "match_order" ] = (df_sorted.set_index( "match_time" ).groupby( "anon_player_id" ).cumcount() + 1).values

#creating match before column to create a merging index
df_sorted[ "match_before" ] = df_sorted[ "match_order" ] - 1

#using merge to put the "idle_gap_min" column from the previous match chronologically into the present
#should have minutes from previous match to current and minutes from current to next in the same column now
temp_df = pd.merge( df_sorted, df_sorted[["anon_player_id", "idle_gap_min", "match_order"]], 
                   left_on = ["anon_player_id","match_before"], 
                   right_on = ["anon_player_id", "match_order"], how = "left",
                   suffixes=('_match_current', '_match_before')
                   ).drop("match_order_match_before", axis = 1 ).rename( {"match_order_match_current" : "match_order",
                                                                         "idle_gap_min_match_current" : "idle_gap_min" }, axis = 1 )

#fill empty values in the minute columns for easier processing
temp_df = temp_df.fillna({
    "idle_gap_min_match_before": -1, 
    "idle_gap_min": -1 })
```

```python
#assignment function for categorical variable to be predicting
# "end series" means match within 45 before but no match within 45 after
# "independent match" means no match within 45 before or after
# "in series" means match within 45 after
def serialization(row):
    before = row["idle_gap_min_match_before"]
    current = row["idle_gap_min"]
    if int(round(current)) == -1 and int(round(before)) < 45:
        return "end series"
    elif int(round(current)) == -1 and int(round(before)) >= 45:
        return "independent match"
    elif int(round(current)) >= 45 and int(round(before)) == -1:
        return "independent match"
    elif int(round(current)) < 45:
        return "in series"
    elif int(round(current)) >= 45 and int(round(before)) < 45:
        return "end series"
    elif int(round(current)) >= 45 and int(round(before)) >= 45:
        return "independent match"
    else:
        return np.NAN

#create categorical column using previous function
temp_df["series_categorical"] = temp_df.apply( serialization, axis = 1 )

#make sure no failures to assign
print(temp_df["series_categorical"].isna().sum())

# save processed reduced dataset
os.makedirs('./data/02-processed', exist_ok=True)
temp_df.to_csv('./data/02-processed/marvel_rivals_session_temp_df.csv', index=False)
temp_df.to_json('./data/02-processed/marvel_rivals_session_temp_df.json', orient='records', indent=4)
```

We'll plot another graph showing probability rather than proportion to see if this difference is significant. 

```python
# remove independent matches
df_sessions = temp_df[temp_df["series_categorical"] != "independent match"].copy()

# did the match end the session?
df_sessions["ended_session"] = df_sessions["series_categorical"] == "end series"

# compute probability session ends after win vs loss
result = (
    df_sessions.groupby("is_win")["ended_session"]
    .mean()
    .reset_index()
)

# plot
plt.figure(figsize=(6,5))
sns.barplot(data=result, x="is_win", y="ended_session")

plt.xlabel("Match Outcome")
plt.ylabel("Probability Session Ends")
plt.title("Probability a Session Ends After a Win vs Loss")

plt.xticks([0,1], ["Loss", "Win"])
plt.show()
```

```python
from scipy.stats import chi2_contingency

# contingency table
ct = pd.crosstab(df_sessions["is_win"], df_sessions["ended_session"])

# probabilities
prob_end = ct.div(ct.sum(axis=1), axis=0)

# difference in probability
diff_prob = prob_end.loc[False, True] - prob_end.loc[True, True]

# odds ratio
odds_ratio = (ct.loc[False, True] / ct.loc[False, False]) / (ct.loc[True, True] / ct.loc[True, False])

# phi correlation
phi = df_sessions["is_win"].corr(df_sessions["ended_session"])

# chi-square test
chi2, p, _, _ = chi2_contingency(ct)

# summary table
summary = pd.DataFrame({
    "Metric": [
        "P(Session Ends | Loss)",
        "P(Session Ends | Win)",
        "Probability Difference (Loss - Win)",
        "Chi-square p-value"
    ],
    "Value": [
        prob_end.loc[False, True],
        prob_end.loc[True, True],
        diff_prob,
        p
    ]
})

display(summary)
```

Sessions ended after **15.9% of losses** compared to **13.6% of wins**, representing a **2.3 percentage point increase** in the probability that a player stops playing following a loss. Although the difference is relatively small, this pattern suggests that players are slightly more likely to terminate a play session after experiencing a negative outcome. 

A chi-square test indicates that the relationship between match outcome and session termination is statistically significant (p < 0.001), suggesting that the observed difference is unlikely to be due to random variation. However, given the large sample size of the dataset, even small effects can achieve statistical significance. 

The observed 2.3 percentage point effect size therefore provides a more meaningful indicator of the strength of this relationship, which appears modest but consistent with behavioral patterns such as loss aversion.

#### Section 4 of EDA  - KDA Performance and Session Behavior

To examine whether individual performance influences session behavior, we construct a role-adjusted performance score by standardizing key gameplay metrics within each hero role. This allowed performance to be compared across roles using a unified scale.

```python
temp_df["hero_role"].value_counts()
```

Unknown roles will not be included in this analysis.

```python
temp_df.columns
```

```python
from scipy.stats import zscore

metrics = [
    "hero_heal",
    "assists",
    "hero_damage",
    "kills",
    "damage_taken",
    "deaths"
]

for metric in metrics:
    temp_df[f"{metric}_z"] = temp_df.groupby("hero_role")[metric].transform(zscore)

temp_df["deaths_z"] = -temp_df["deaths_z"]
```

Performance statistics are Z-Score normalized since these are the top players and tend to perform much better than average players. Because matchwide statistics were unavailable, performance was normalized within hero roles rather than relative to other players within the same match. While this prevents direct comparison to match-level averages, role-based standardization allows performance to be interpreted relative to typical gameplay expectations for each role.

```python
temp_df.columns
```

```python
temp_df["support_score"] = (
    temp_df["hero_heal_z"] +
    temp_df["assists_z"] +
    temp_df["deaths_z"] +
    temp_df["kills_z"]
)

temp_df["dps_score"] = (
    temp_df["hero_damage_z"] +
    temp_df["kills_z"]
)

temp_df["tank_score"] = (
    temp_df["damage_taken_z"] +
    temp_df["kills_z"] +
    temp_df["hero_damage_z"] +
    temp_df["deaths_z"]
)

def performance_score(row):
    if row["hero_role"] == "Strategist":
        return row["support_score"]
    elif row["hero_role"] == "Duelist":
        return row["dps_score"]
    elif row["hero_role"] == "Vanguard":
        return row["tank_score"]

#drop irrelevant scores
temp_df = temp_df.drop(columns=["support_score","dps_score","tank_score", 'match_order', 'match_before', 'hero_heal_z', 'assists_z',
       'hero_damage_z', 'kills_z', 'damage_taken_z', 'deaths_z'])

os.makedirs('./data/02-processed', exist_ok=True)
temp_df.to_json('./data/02-processed/marvel_rivals_final.json', orient='records', indent=4)
temp_df.to_csv('./data/02-processed/marvel_rivals_final.csv', index=False)
```

We construct a role-adjusted performance score for each match. First, it calculates separate performance metrics for supports (Strategists), DPS (Duelists), and tanks (Vanguards) by summing standardized (z-scored) gameplay statistics that are most relevant to each role, such as healing and assists for supports or damage and kills for DPS. It then assigns the appropriate role-specific score to each row based on the player’s hero role, producing a single performance_score that represents overall performance relative to typical expectations for that role.

```python
temp_df.groupby("series_categorical")["performance_score"].describe()
```

Summary statistics show that matches occurring at the end of a play session exhibit slightly lower average performance (mean = −0.012) compared to matches that occur within an ongoing session (mean = −0.008). Median values show a similar pattern, suggesting that players tend to perform marginally worse in the final match before stopping.

Interestingly, independent matches—sessions consisting of only a single game—show a noticeably higher mean performance score (mean = 0.27). This may indicate that some players stop playing after achieving a satisfactory outcome rather than due to frustration.

Overall, while the differences are modest, the results suggest that performance may play a minor role in session termination behavior, with players slightly more likely to end sessions following below-average performance.

**However**, another question arises: **How strongly do performance scores predict match outcome?** Because if high performance and match outcome are 1-1 correlated, then we only need to look at match outcome to determine whether it has an effect on session behavior. 

```python
temp_df.groupby("is_win")["performance_score"].describe()
```

The summary statistics show a clear difference in performance scores between winning and losing matches.

Losses: mean = -1.04, median = -1.322

Wins: mean = 0.694, median = 0.418

This indicates that players tend to have higher performance scores in matches they win and lower scores in matches they lose, which is expected since the performance score includes metrics like kills, damage, healing, and deaths that contribute to overall match success.

However, the distributions overlap substantially, as shown by the wide ranges and standard deviations (≈2.0 for both groups). This suggests that individual performance does not perfectly determine match outcome.

Next we calculate the Pearson Correlation Coefficient:

```python
temp_df["win_binary"] = temp_df["is_win"].map({True:1, False:0})

temp_df["performance_score"].corr(temp_df["win_binary"])
```

Performance score is moderately correlated with match outcome (r = 0.37), indicating that stronger individual performance increases the likelihood of winning but does not fully determine match results in this team-based game.

The fact that correlation is only moderate at best raises another question: What are  other factors that lead a player towards a win if their performance was only average? Some weaker players win despite low performance, which would be a "carried win." Some stronger players lose despite high performance, which would be a "frustrating loss." There must be enough of these mismatched cases occuring to weaken the correlation of performance and match outcome.

```python
temp_df["performance_level"] = temp_df["performance_score"].apply(
    lambda x: "high" if x > 0 else "low"
)

def performance_outcome(row):
    
    if row["performance_level"] == "high" and row["is_win"] == True:
        return "high performance + win"
    
    elif row["performance_level"] == "low" and row["is_win"] == False:
        return "low performance + loss"
    
    elif row["performance_level"] == "high" and row["is_win"] == False:
        return "high performance + loss"
    
    else:
        return "low performance + win"

temp_df["performance_outcome"] = temp_df.apply(performance_outcome, axis=1)

temp_df["performance_outcome"].value_counts(normalize=True)

pd.crosstab(
    temp_df["performance_outcome"],
    temp_df["series_categorical"],
    normalize="index"
)
```

Matches were categorized into four groups based on whether player performance was above or below average and whether the match resulted in a win or loss. The distribution of session types across these categories is highly similar, with approximately 82–84% of matches occurring within an ongoing series and only 13–16% representing session-ending matches. Among these groups, low performance + loss has the largest proportion of session-ending matches (15.7%), though it differs only slightly from high performance + loss (15.0%). 

```python
# compute probability that a match ends the session
quit_rates = (
    temp_df.assign(end_series = temp_df["series_categorical"] == "end series")
    .groupby("performance_outcome")["end_series"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(8,5))

sns.barplot(
    data=quit_rates,
    x="performance_outcome",
    y="end_series"
)

plt.ylabel("Probability of Ending Session")
plt.xlabel("Performance–Outcome Category")
plt.title("Session Termination by Performance–Outcome Category")
plt.xticks(rotation=30)

plt.show()
```
